
### Ejercicio 3

Completa el cdigo para poder reproducir la imagen de la figura, tendrs que revisar el parmetro "connectionstyle" del rango de parmetros del argumento "arrowprops". 
Nota: Las coordenadas del texo no estn dadas como puntos de la grfica sino como "offset" del punto sealado. Esto quiere decir que si el offset es la tupla (-2,10) y las coordenadas (xy) del punto sealado por la anotacin son (30,40), por ejemplo, el texto se situar en (30-2, 40 + 10) es decir en (28,50) [Recuerda, es la coordenada de la esquina inferior izquierda del cuadro de texto "invisible" que contiene al texto]

```python
from matplotlib.patches import ConnectionStyle, FancyArrowPatch
```

```python

plt.plot(x, y, marker='o')
coordenadas = (3,5)
angulo = ConnectionStyle.Angle3(angleA = 0, angleB = 90)

props_flecha = {
    "arrowstyle": "->",
    "connectionstyle": angulo}

plt.annotate("Punto destacado", xy =coordenadas, textcoords='offset points', xytext=(-25, 40),
             arrowprops=props_flecha)

plt.title('Anotacin Bsica')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

```


### Ejercicio 4

Siguiendo el estilo de offset del ejercicio anterior, completa el cdigo siguiente para poder reproducir la imagen de la figura:




```python
plt.plot(x, y, marker='o')

props_flecha_x = {
    "arrowstyle": "->",
    "connectionstyle": "arc",
    "color": "black",
    "lw": 5}

props_flecha_y = {
    "arrowstyle": "->",
    "connectionstyle": "arc",
    "color": "red",
    "lw": 5,
    "edgecolor": "black",
    }
plt.annotate("Inicio", xy =(1,2), textcoords='offset points', xytext=(100, 2),
             arrowprops=props_flecha_x)

plt.annotate("Final", xy =(5,11), textcoords='offset points', xytext=(-50, -10),
             arrowprops=props_flecha_y)

plt.title('Ejercicio 4')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### Ejercicio 5

Carga el dataset de accidentes de trfico en EEUU que est en la ruta "./data/car_crashes.csv". Quedate con los cinco estados con menos accidentes en "total".

```python

df_crash = pd.read_csv("./data/car_crashes.csv", sep = "|")
df_crash.head()
```

```python
total_crash = df_crash.groupby("abbrev")["total"].sum()

crash_sorted = total_crash.sort_values()

less_crash = crash_sorted.head(5)

less_crash

df_top_min = pd.DataFrame(less_crash)

df_top_min.reset_index(inplace = True)
df_top_min
```

### Ejercicio 6

Completa el siguiente cdigo para obtener una figura como la representada en la imagen

```python
#OJO df_top_min contiene el subdataframe resultado del ejercicio anterior
estados = {
    "DC": "District of Columbia",
    "MA": "Massachusetts",
    "MN": "Minnesota",
    "WA": "Washington",
    "CT": "Connecticut"
}

for abrev,estado in estados.items():
    df_top_min["abbrev"] = df_top_min["abbrev"].str.replace(abrev,estado)

colores = ["green","gray","orange","blue","pink"]
plt.figure(figsize = (10,5))
plt.bar(df_top_min["abbrev"], df_top_min["total"], color = colores, alpha = 0.5)
plt.ylabel("Total accidenres por millones de millas recorridas")
plt.ylim(0,30)
plt.axhline(y = df_crash.total.max());
plt.title("Ejercicio 6");
```

### Ejercicio 7

Copia el cdigo (una vez lo hayas completado) del ejercicio anterior en una celda nueva, aade las lneas necesarias para reconstruir la siguiente figura:

```python
estados = {
    "DC": "District of Columbia",
    "MA": "Massachusetts",
    "MN": "Minnesota",
    "WA": "Washington",
    "CT": "Connecticut"
}

for abrev,estado in estados.items():
    df_top_min["abbrev"] = df_top_min["abbrev"].str.replace(abrev,estado)

colores = ["green","gray","orange","blue","pink"]
plt.figure(figsize = (10,5))
bars = plt.bar(df_top_min["abbrev"], df_top_min["total"], color = colores, alpha = 0.5, )

for bar in bars:
    yval = bar.get_height() # devuelve la altura de la cada barra al iterar sobre ellas
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, round(yval, 1), ha='center', va='bottom')

index_barra = 0  # ndice de la barra a destacar
plt.annotate('DC', xy=(index_barra, (df_top_min["total"].iloc[index_barra]+2)),
             xytext=(index_barra, df_top_min["total"].iloc[index_barra] + 5),
             arrowprops=dict(facecolor='black', arrowstyle='<->'),
             ha='center', va='bottom')

plt.ylabel("Total accidenres por millones de millas recorridas")
plt.ylim(0,30)
plt.axhline(y = df_crash.total.max());
plt.text(2, df_crash.total.max() + 1, "Mximo valor para todos los estados", ha = "center")
plt.title("Ejercicio 6");
```
